```python
# block tables
# Overlap blocking 
ob = em.OverlapBlocker()

# block using name
C0 = ob.block_tables(A,B, 'Directors', 'Directors', l_output_attrs=['Title', 'Directors'],r_output_attrs=['Title', 'Directors'],word_level=True, overlap_size=1)
```

```python
C0
```

```python
# Overlap blocking over candidata set (C0)
ob = em.OverlapBlocker()

# block using name 
C1 = ob.block_candset(C0, 'Title', 'Title', word_level=True, overlap_size=1)
```
